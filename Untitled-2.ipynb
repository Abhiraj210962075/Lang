```python
import librosa
import numpy as np
import matplotlib.pyplot as plt
from difflib import SequenceMatcher
import whisper
from gtts import gTTS

# Load Whisper AI model once
whisper_model = whisper.load_model("base")

def load_and_preprocess(audio_path):
    # Load audio file
    audio, sr = librosa.load(audio_path, sr=None)
    # Normalize audio
    audio = librosa.util.normalize(audio)
    return audio, sr

def plot_waveforms(master_audio, user_audio, master_sr, user_sr):
    plt.figure(figsize=(12, 6))
    plt.subplot(2, 1, 1)
    librosa.display.waveshow(master_audio, sr=master_sr)
    plt.title("Master Audio Waveform")
    plt.subplot(2, 1, 2)
    librosa.display.waveshow(user_audio, sr=user_sr)
    plt.title("User Audio Waveform")
    plt.tight_layout()
    plt.show()

def transcribe_with_whisper(audio_path, language="ar"):
    result = whisper_model.transcribe(audio_path, language=language)
    return result

def compare_transcriptions(master_transcription, user_transcription):
    matcher = SequenceMatcher(None, master_transcription, user_transcription)
    similarity = matcher.ratio()
    return similarity

def find_mistakes(master_transcription, user_transcription):
    matcher = SequenceMatcher(None, master_transcription, user_transcription)
    diff = list(matcher.get_opcodes())

    mistakes = []
    for tag, i1, i2, j1, j2 in diff:
        if tag != 'equal':
            mistakes.append((master_transcription[i1:i2], user_transcription[j1:j2], j1, j2))
    return mistakes

def get_timestamped_mistakes(transcription_result, mistakes):
    timestamped_mistakes = []
    segments = transcription_result['segments']
    for master, user, j1, j2 in mistakes:
        if not user.strip():
            continue  # Skip empty user segments
        for segment in segments:
            start_time = segment['start']
            end_time = segment['end']
            text = segment['text']
            if text.find(user) != -1:
                timestamped_mistakes.append((start_time, end_time, master, user))
                break
    return timestamped_mistakes

def speak_mistakes(timestamped_mistakes):
    if not timestamped_mistakes:
        print("No mistakes found.")
        return
    for i, (start_time, end_time, master, user) in enumerate(timestamped_mistakes, start=1):
        mistake_text = f"From {start_time:.2f}s to {end_time:.2f}s: Expected '{master}', but heard '{user}'"
        print(mistake_text)
        if user.strip():  # Ensure the user text is not empty
            tts = gTTS(text=user, lang='ar')
            tts.save(f"mistake_{i}.mp3")

def evaluate_recitation(master_audio_path, user_audio_path):
    # Load and preprocess audios
    master_audio, master_sr = load_and_preprocess(master_audio_path)
    user_audio, user_sr = load_and_preprocess(user_audio_path)
    
    # Plot waveforms for visual analysis
    plot_waveforms(master_audio, user_audio, master_sr, user_sr)
    
    # Transcribe master audio
    master_transcription_result = transcribe_with_whisper(master_audio_path)
    master_transcription = master_transcription_result['text']
    print(f"Master Transcription: {master_transcription}")
    
    # Transcribe user audio
    user_transcription_result = transcribe_with_whisper(user_audio_path)
    user_transcription = user_transcription_result['text']
    
    # Compare transcriptions
    similarity = compare_transcriptions(master_transcription, user_transcription)
    
    # Find mistakes made by the user
    mistakes = find_mistakes(master_transcription, user_transcription)
    
    # Get timestamped mistakes
    timestamped_mistakes = get_timestamped_mistakes(user_transcription_result, mistakes)
    
    # Speak mistakes
    speak_mistakes(timestamped_mistakes)
    
    return similarity

# Example usage
master_audio_path = "/Users/abhirajtayer/Desktop/AI INTERNSHIP/TestAudioSet/Surah_Ikhlaas_Master.wav"
user_audio_path = "/Users/abhirajtayer/Desktop/AI INTERNSHIP/TestAudioSet/Surah_Ikhlaas_Good_Recording.wav"

accuracy_score = evaluate_recitation(master_audio_path, user_audio_path)
print(f"Recitation Accuracy Score: {accuracy_score:.2f}")
```
